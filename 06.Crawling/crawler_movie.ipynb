```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
from html import unescape
import requests
import re
```

```python
def fetch(url):  
    f = urlopen(url) # url에 해당하는 web page를 통으로 크롤링
    
    # HTTP 헤더를 기반으로 인코딩 형식을 추출
    encoding = f.info().get_content_charset(failobj="utf-8")
    print("--- 1. 인코딩 정보 : ", encoding)    # utf-8

    
    # 추출한 인코딩 형식을 기반으로 문자열을 디코딩, 정보로 디코딩 안 할 경우 한글 깨짐 현상이 발생 가능
    # 왜? 특정사이트는 encoding 정보가 utf-8으로 되어 있기도 함, 해당 사이트가 구현되어있는 정보로 디코딩 수행
    # f 변수가 보유한 페이지의 모든 정보를 utf-8 포멧으로 복원 
    html = f.read().decode(encoding)

    #print("--- 2. 모든 문서 내용 추출 ", html)
    return html

url = 'https://movie.naver.com/movie/running/current.nhn#'

response = requests.get(url)

if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    print(soup)
```

```python
# 네이버 검색창에서 뽑아오기

url = 'https://search.naver.com/search.naver?ie=UTF-8&sm=whl_hty&query=%EB%84%A4%EC%9D%B4%EB%B2%84%EC%98%81%ED%99%94'

response = requests.get(url)

if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    print(soup)
```

```python
movie = []

for data in soup.select('div.data_box > div > div > a'):
#     print(data['href'])
#     print(data.string)
    title = data.string
    url = 'https://search.naver.com/search.naver' + data['href']
    
for info in soup.select('div.info > dl > dd > span.num'):
# for info in soup.select('div.info > dl'):
#     print(info.string)
#     print(info.dd.string)
    score = info.string
    
    
    movie.append({'title':title, 'url':url, 'score':score})
print(movie)
```

```python
# 다음영화에서 뽑아오기

url = 'https://movie.daum.net/ranking/reservation'

response = requests.get(url)

if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    print(soup)
```

```python
movie = []

for data in soup.select('div.item_poster > div > strong > a'):
#     print(data)
#     print(data['href'])
#     print(data.string)
    title = data.string
    url = 'https://movie.daum.net' + data['href']
    movie.append({'title':title, 'url':url})
    
for info in soup.select('div.item_poster > div > span > span > span'):
    if re.search(r'\d$', info.string):
#         print(info.string)
        score = info.string
    if re.search(r'\d%$', info.string):
#         print(info.string)
        advance_rate = info.string
#         print(advance_rate)
        
                 
    
    
    movie.append({'title':title, 'url':url, 'score':score, 'advance_rate':advance_rate})
print(movie)
```
