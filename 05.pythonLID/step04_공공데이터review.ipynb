```python
c=['무너짐','넘어짐','화재','없음','떨어짐','낙하','파괴','기울어짐','붕락','붕괴']
d=df['물적피해'].value_counts()
plt.figure(figsize=(10,10)) ## Figure 생성 사이즈는 10 by 10
# xtick_label_position = list(range(len(age_category))) ## x축 눈금 라벨이 표시될 x좌표
# plt.xticks(xtick_label_position, age_category) ## x축 눈금 라벨 출력
## 바 차트 출력, 막대기 색깔을 초록색으로 설정
plt.bar(c,d, color='green')
plt.title('물적피해 카운트',fontsize=20) ## 타이틀 출력
plt.xlabel('물적피해') ## x축 라벨 출력
plt.ylabel('숫자') ## y축 라벨 출력
```

```python
damage = df.groupby('물적피해')
df4 = pd.DataFrame({'count' : damage.size()}).reset_index().sort_values(by='count', ascending=False)
df4
```

```python
# df3 = titanic.pclass.value_counts()
# 파이차트 : https://wikidocs.net/92114 
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 2}
plt.figure(figsize=(8,8))
plt.pie(df4['count'], labels= df4['물적피해'], autopct='%.1f%%', startangle=200, counterclock=False, wedgeprops=wedgeprops)
plt.title("물적피해")
plt.axis('equal')
plt.show()
```

```python
def get_year(row): # 'yyyy-mm-dd' 포멧으로 parameter가 유입된다 가정
    
    return str(row)[0:4]

df['year'] = df['구분'].apply(get_year)
df
```

```python
d=df['year'].value_counts()
d
```

```python
year = df.groupby('year')
df2 = pd.DataFrame({'count' : year.size()}).reset_index()
df2
```

```python
df3 = df2.sort_values(by='year', ascending=False)
df3
```

```python

plt.figure(figsize=(5,5)) ## Figure 생성 사이즈는 10 by 10
# xtick_label_position = list(range(len(age_category))) ## x축 눈금 라벨이 표시될 x좌표
# plt.xticks(xtick_label_position, age_category) ## x축 눈금 라벨 출력
## 바 차트 출력, 막대기 색깔을 초록색으로 설정
plt.bar(df3['year'], df3['count'], color='Orchid')
plt.title('위험요소 카운트',fontsize=20) ## 타이틀 출력
plt.xlabel('년도') ## x축 라벨 출력
plt.ylabel('횟수') ## y축 라벨 출력
```
