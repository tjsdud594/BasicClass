```python
# 레스토랑 팁 데이터 셋 활용
'''
미션 : tip을 어떤 사람이 많이 주는지?
분석 - 데이터 취합

'''

tips = sns.load_dataset('tips')
tips
```

```python
tips.head(10)
```

```python
tips.columns
```

```python
# total_bill 즉 계산한 금액과 tip 관계도 확인 - 산점도
# ; 찍으면 chart 위 <seaborn.axisgrid.FacetGrid object at 0x00000196396B9970> 안나옴
print(sns.regplot(x='total_bill', y='tip', data=tips))
sns.relplot(x='total_bill', y='tip', data=tips);
```

```python
# 스타일 적용
sns.set_style(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips);
```

```python
# 스타일 적용, feature 를 색으로 구분
# 흡연 비흡연에 대한 색을 구분 - 범례표현도 자동설정()

sns.set_style(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker');
```

```python
sns.set_style(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue='size');
```

```python
# hue는 색상차이로 구분, style은 모양으로 구분

sns.set_style(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', style='smoker');
```

```python
sns.set_style(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', style='smoker', palette='Blues');
```
