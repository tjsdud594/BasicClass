# Image란?

## Image는 width X height X channel 의 pixels로 이루어진 matrix

- width: 너비로 이미지의 가로 길이를 의미한다.
- height: 높이로 이미지의 세로 길이를 의미한다.

- pixel: image가 가지고 있는 값으로 width X height 개수만큼 존재한다. image를 표현하는 bit방식에 따라 값의 범위가 달라진다.

- 일반적인 Image는 unsigned integer 8bit로 표현
    - 0 ~ 255 로 표현
- bit수가 높아질 수록 표현할 수 있는 색의 범위가 증가한다.

## 색공간 변환
- **`cv2.cvtColor(src, code)`**
    - image 의 color space를 변환한다. 
    - src: 변환시킬 이미지 (ndarray)
    - code
        - 변환시킬 색공간 타입 지정
        - cv2.COLOR_XXX**2**YYY 형태의 상수 지정(XXX를YYY로 변환)
            - cv2.COLOR_XXX2YYY 형태의 상수 지정(XXX를YYY로 변환)
            - cv2.COLOR_BGR2GRAY / cv2.COLOR_GRAY2BGR (BGR <-> GRAY)
            - cv2.COLOR_BGR2RGB  / cv2.COLOR_RGB2BGR  (BGR <-> RGB)
            - cv2.COLOR_BGR2HSV  / cv2.COLOR_HSV2BGR  (BGR <-> HSV)

> ### HSV
> - Hue: 색상, 색의 종류
> - Saturation: 채도. 색의 선명도
> - Value : 명도. 밝기

# 동영상 처리

## 동영상 읽기
- VideoCapture 클래스사용
    - 객체 생성
        - VideoCapture('동영상파일 경로'): 동영상파일
        - VideoCapture(웹캠 ID): 웹캠
- VideoCapture의 주요 메소드
    - isOpened(): bool
        - 입력 대상과 연결되었는지 여부반환
    - read(): (bool, img)
        - Frame 이미지로 읽기
        - 반환값
            - bool: 읽었는지 여부
            - img: 읽은 이미지(ndarray)

### 웹캠

```python
import cv2
import numpy as np


cap = cv2.VideoCapture(0)  # 정수 : 연결하려는 webcam id  (0 ~), 문자열 : 동영상 파일경로

# 연결 성공 여부 확인
if cap.isOpened()==False:  # 연결되었으면 True
    print('웹캠 연결 실패')
    exit(1)  # exit(종료코드:정수) : 프로그램 종료 (0-정상종료, 양수/음수-비정상종료)
    
# 웹캠으로부터 영상을 읽기
# 웹캠에서 입력되는 이미지들을 캡쳐해서 회면(imshow())에 보여준다.
while True:
    # 한장의 이미지를 캡처 (한번 반복시마다 한장씩 캡쳐)
    ret, img = cap.read()   # ret: 읽었는지 여부, img: 이미지 배열(ndarray)
    
    img = cv2.flip(img, 1)
    # flip(이미지배열, 모드) : 반전된 배열 반환
    # 양수 - 좌우반전(수평)
    # 0 - 상하반전(수직)
    # 음수 - 좌우상하반전
    
    if ret==False:
        print('캡처 실패')
        break  # 캡처종료
        
    cv2.imshow('frame', img)
    if cv2.waitKey(1)==27:  # ESC 키 클릭시 종료
        break
        
        
# end of while
cap.release() # 웹캠 연결을 닫아준다.
cv2.destroyAllWindows()
```

## TODO : 웹캠 영상을 이미지로 저장
- output_capture / cap_번호.jpg

    1. 웹캠 연결

    2. capture (while)     
        한장캡쳐    
        window에 출력      
        캡쳐된 이미지를 파일로 저장 - 100장당 한장씩만 저장     
        esc 키입력되면 종료

```python
cap = cv2.VideoCapture(0)  # 정수 : 연결하려는 webcam id  (0 ~), 문자열 : 동영상 파일경로

# 연결 성공 여부 확인
if cap.isOpened()==False:  # 연결되었으면 True
    print('웹캠 연결 실패')
    exit(1)  # exit(종료코드:정수) : 프로그램 종료 (0-정상종료, 양수/음수-비정상종료)
    
# 웹캠으로부터 영상을 읽기
# 웹캠에서 입력되는 이미지들을 캡쳐해서 회면(imshow())에 보여준다.

base_dir = 'output_capture'
count = 0
while True:
    # 한장의 이미지를 캡처 (한번 반복시마다 한장씩 캡쳐)
    ret, img = cap.read()   # ret: 읽었는지 여부, img: 이미지 배열(ndarray)
    
    img = cv2.flip(img, 1)
    # flip(이미지배열, 모드) : 반전된 배열 반환
    # 양수 - 좌우반전(수평)
    # 0 - 상하반전(수직)
    # 음수 - 좌우상하반전
    count += 1
    if ret==False:
        print('캡처 실패')
        break  # 캡처종료
        
    cv2.imshow('frame', img)
    if cv2.waitKey(1)==27:  # ESC 키 클릭시 종료
        break
        
    if not os.path.isdir(base_dir):  # output 디렉토리가 없다면
        os.mkdir(base_dir)
        
    if count % 100 == 0:
        cv2.imwrite('output_capture/cap_'+str(count)+'.jpg', img)
        
# end of while
cap.release() # 웹캠 연결을 닫아준다.
cv2.destroyAllWindows()
```

```python
# 강사님 버전!!
import cv2
import os

# 파일을 저장할 디렉토리 생성
base_dir = 'output_capture'
if not os.path.isdir(base_dir):
    os.mkdir(base_dir)
    
# webcam 연결
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print('웹캠 연결 실패')
    exit(1)
    
# 캡쳐작업
cnt = 1 # 파일명 뒤에 붙여줄 번호
while True:
    
    ret, frame = cap.read()
    if not ret:
        print('캡쳐 실패')
        break
        
    # 반전
    frame = cv2.flip(frame, 1)
    
    # 화면 출력
    cv2.imshow('frame', frame)
    
    # 파일로 저장
    if cnt % 100 == 0:
        file_path = os.path.join(base_dir, f'cap_{cnt}.jpg')
        cv2.imwrite(file_path, frame)
    cnt+=1
    
    if cv2.waitKey(1)==27:
        break
        
cap.release()
cv2.destroyAllWindows()
```

**동영상 파일 읽기**

- FPS : Frame Per Second
    - 동영상이 1초에 몇장의 frame(이미지)를 보여주는지를 알려주는 수치
    - ex) 30FPS : 1초에 30장의 이미지를 보여줌

```python
import cv2
import os

cap = cv2.VideoCapture('images/wave.mp4')  # 동영상 파일과 연결 -> 동영상 파일 경로

if not cap.isOpened():
    print('동영상 로딩 실패')
    exit(1)
    
# FPS 조회
# VideoCapture객체.get(상수) : 연결된 동영상 소스로부터 상수(CAP_XXXX)와 관련된 정보를 반환
FPS = cap.get(cv2.CAP_PROP_FPS) 
print(FPS)
DELAY = round(1000/FPS)  # 얼마에 한번씩 캡처를할지 시간 계산,정수로 만들어 주기 위해 반올림 처리.
    
# 읽기
while True:
    
    ret, frame = cap.read()
    if not ret:
        print('동영상 캡처 실패')
        break
        
    # 캡쳐된 영상을 화면에 출력
    cv2.imshow('wave', frame)
    
    if cv2.waitKey(DELAY)==27:
        break
        
cap.release()
cv2.destroyAllWindows()
```

### 동영상 저장
- capture(read)한 이미지 프레임을 연속적으로 저장하면 동영상 저장이 된다.
- VideoWriter 객체를 이용해 저장
    - **`VideoWriter(filename, codec, fps, size)`**
        - filename: 저장경로
        - codec
            - VideoWriter_fourcc 이용
            
        - fps
            - FPS (Frame Per Second) - 초당 몇 프레임인지 지정
        - size
            - 저장할 frame 크기로 원본 동영상이나 웹캠의 width, height 순서로 넣는다.
    - `VideoWriter().write(img)`
        - Frame 저장

```python
# 웹캠에서 캡처한 프레임(이미지영상)을 동영상 파일로 저장.
import os 
import cv2

cap = cv2.VideoCapture(0) 
if not cap.isOpened():
    print('웹캠 연결 실패')
    exit(1)
    
# frame 1장 read() -> 캡쳐된 이미지 size확인을 위함
ret, img = cap.read()
if not ret:
    print('캡처 실패')
    exit(1)
    
print(img.shape)
height = img.shape[0]
width = img.shape[1]

# Webcam FPS
FPS = cap.get(cv2.CAP_PROP_FPS)
print(f'FPS : {FPS}')

# Codec
codec = cv2.VideoWriter_fourcc(*'MJPG')   # ['M','J','P','G']

writer = cv2.VideoWriter('output/web_cam_output.avi',  # 저장할 파일경로
                         codec,
                         FPS,
                         (width, height)   # size - (width, height 순서)
                        )

if not writer.isOpened():
    print('동영상을 저장할 수 없습니다')
    exit(1)

# VideoCapture로 frame을 읽어서 VideoWriter로 파일에 출력
while True:
    ret, frame = cap.read()
    if not ret:
        print('캡처 실패')
        break
        
    #화면에 출력
    cv2.imshow('frame',frame)
    #파일에 저장
    writer.write(frame)
    
    if cv2.waitKey(1)==27:
        break
    
    
    
    
cap.release()
writer.release()
cv2.destroyAllWindows()
```
